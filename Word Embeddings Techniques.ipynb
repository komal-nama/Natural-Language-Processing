```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
# sentences (every sentence has diff length)
sent=['the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good']
```

```python
# Vocabulary size (large vocabulary size, good amt of feature representation) 
voc_size=500
```

#### One Hot Representation 

```python
onehot_repr = [one_hot(words,voc_size)for words in sent] 
print(onehot_repr)

#O/P - index no. where it is one 
```

#### Word Embedding Represntation

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
# Pre padding   (Ensures that every sentence will have length=8 (either add 0 at beginning or last))
sent_length = 8    #maximum sentence legth = 8
embedded_docs = pad_sequences(onehot_repr, padding = 'pre', maxlen = sent_length)  #pre = 0 in forward
print(embedded_docs)
```

```python
#now input size is fixed
```

```python
# Post padding (0 in last)
embedded_docs = pad_sequences(onehot_repr, padding = 'post', maxlen = sent_length)  #pre = 0 in forward
print(embedded_docs)
```

```python
dim = 10    #(for each feature, 10 dimensions[]
```

```python
model = Sequential()
model.add(Embedding(voc_size, 10, input_length = sent_length))   #Embedding layer will work similar to Word2Vec
# 10 = no. of features for each vector
model.compile('adam', 'mse') 
```

```python
model.summary()
```

```python
# 'the glass of milk'
embedded_docs[0]
```

```python
model.predict(embedded_docs[0])    #1st sentence
```
